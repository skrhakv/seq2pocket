```python

stats = [
    (0.53, 'MeanShift with bandwidth=9', 2.14),
    (0.23, 'MeanShift with bandwidth=12', 1.19),
    (0.66, 'DBSCAN with eps=5', 4.83),
    (0.46, 'DBSCAN with eps=7', 1.54),
    (0.36, 'GMM', 1.52),
    (0.69, 'Our Method', 1.9),
]

import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6, 6))
plt.title('Clustering Method Comparison')
plt.ylabel('DCC (Higher is Better)')
plt.xlabel('PFI (Lower is Better)')

for DCC, name, PFI in stats:
    plt.scatter(PFI, DCC, s=100)
    if name != 'DBSCAN with eps=5':
        plt.text(PFI + 0.02, DCC + 0.02, name, fontsize=9)
    else:
        plt.text(PFI - 1.2, DCC + 0.02, name, fontsize=9)
plt.xlim(0, 5)
plt.ylim(0, 1)
plt.grid(True)
plt.show()    
```
