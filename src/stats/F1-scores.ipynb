```python
import pickle

DATA_PATH = '/home/vit/Projects/cryptoshow-analysis/data/stats'
with open(f'{DATA_PATH}/LIGYSIS_predictions.pkl', 'rb') as file:
    df = pickle.load(file)
df.head()
```

```python
from sklearn.metrics import f1_score, accuracy_score

actual = df['LIGYSIS'].to_numpy()
predicted_p2rank = df['P2Rank+Cons'].to_numpy()
predicted_puresnet = df['PUResNet'].to_numpy()
predicted_ligsite = df['Ligsite'].to_numpy()
predicted_fpocket = df['fpocket'].to_numpy()

predictors = {
    'P2Rank': predicted_p2rank,
    'PUResNet': predicted_puresnet,
    'Ligsite': predicted_ligsite,
    'fpocket': predicted_fpocket
}

for name, pred in predictors.items():
    acc = accuracy_score(actual, pred)
    f1 = f1_score(actual, pred, average='weighted')
    print(f"{name} - Accuracy: {acc:.4f}, F1: {f1:.4f}")
```
