```python

stats = [
    (0.593, 'MeanShift with bandwidth=9', 2.22),
    (0.272, 'MeanShift with bandwidth=12', 1.19),
    (0.692, 'DBSCAN with eps=5', 4.68),
    (0.464, 'DBSCAN with eps=7', 1.46),
    (0.402, 'GMM', 1.55),
    (0.688, 'Seq2Pocket (Our Method)', 2.12),
]

import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6, 6))
plt.title('Clustering Method Comparison')
plt.ylabel('DCC (Higher is Better)')
plt.xlabel('PFI (Lower is Better)')

for DCC, name, PFI in stats:
    plt.scatter(PFI, DCC, s=100)
    if name == 'Seq2Pocket (Our Method)':
        plt.text(PFI - 0.7, DCC + 0.02, name, fontsize=9)
    elif name != 'DBSCAN with eps=5':
        plt.text(PFI + 0.02, DCC + 0.02, name, fontsize=9)
    else:
        plt.text(PFI - 1.2, DCC + 0.02, name, fontsize=9)
plt.xlim(0, 5)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
  
# plt.savefig('/home/vit/Projects/cryptoshow-analysis/data/stats/clustering-method-comparison.png')
```
