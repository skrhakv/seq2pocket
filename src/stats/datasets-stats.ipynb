```python
dataset_paths = ['/home/vit/Projects/cryptic-nn/data/cryptobench/test.txt',
'/home/vit/Projects/cryptic-nn/data/cryptobench/train.txt',
'/home/vit/Projects/cryptoshow-analysis/data/E-regular-binding-site-predictor/ligysis_without_unobserved.csv',
'/home/vit/Projects/cryptoshow-analysis/data/E-regular-binding-site-predictor/scPDB_enhanced_binding_sites_translated.csv',
'/home/vit/Projects/cryptoshow-analysis/data/E-regular-binding-site-predictor/full_scPDB.csv']

import csv 

for path in dataset_paths:
    protein_ids = []
    binding_site_lens = []
    sequence_lens = []
    with open(path, 'r') as file:
        reader = csv.reader(file, delimiter=';')
        for row in reader:
            protein_id = row[0] + row[1]
            binding_site_len = len(row[3].split(' '))
            sequence_len = len(row[4])
            protein_ids.append(protein_id)
            binding_site_lens.append(binding_site_len)
            sequence_lens.append(sequence_len)
    print(f"Dataset: {path}")
    print(f"Number of unique proteins: {len(set(protein_ids))}")
    print(f"Number of residues in binding sites: {sum(binding_site_lens)}")
    print(f"Number of residues overall: {sum(sequence_lens)}")
    print(f"Percentage of residues in binding sites: {sum(binding_site_lens) / sum(sequence_lens) * 100:.2f}%")
    print() 
```
